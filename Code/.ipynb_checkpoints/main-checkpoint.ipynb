## Import Data & Convert CSV to Parquet

```python
import pandas as pd
```

```python
# Ganti 'nama_file.csv' dengan path file CSV Anda
df = pd.read_csv(r'C:\MyFolder\Perkuliahan\Semester 8\Tugas Akhir\Data\data elisa dari database\data_mentah\data_jam_full.csv', encoding='utf-8')  # atau encoding='latin1' jika ada error karakter

# Tampilkan 5 baris pertama untuk memastikan data terbaca
df.head()
```
